```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import scipy.optimize as opt
```

```python
data=np.loadtxt("ex2data1.txt",delimiter=",")
x_input=data[:,0:2]
y_input=data[:,-1]
m=y_input.shape[0]
n=x_input.shape[1]
```

```python
def plotData(data,x,y):
    data_positive=data[data[:,-1]==1]
    data_negative=data[data[:,-1]==0]
    x_positive=data_positive[:,0:2]
    x_negative=data_negative[:,0:2]
    plt.scatter(x_positive[:,0],x_positive[:,1],color="k",marker="+",label="Admitted")
    plt.scatter(x_negative[:,0],x_negative[:,1],color="y",marker="o",label="Not admitted")
    plt.xlabel("Exam 1 score")
    plt.ylabel("Exam 2 score")
    plt.legend(loc="upper right")       
```

```python
plotData(data,x_input,y_input)
```

```python
# add coloumn of ones to input data for intercept term
x = np.c_[np.ones(m), x_input]
 
```

```python
def sigmoid(z):
    
    h=1/(1+np.exp(-z))
    return h
    
```

```python
def costFunc(x,y,theta):
    z=np.dot(x,theta)
    h=sigmoid(z)
    log_h=np.log(h)
    h_1=np.subtract(1,h)
    log_h_1=np.log(h_1)
    y_1=np.subtract(1,y)
    
    J=np.subtract(-y * log_h, y_1 * log_h_1)
    J=(1/m)*np.sum(J)
    grad=(1/m)*np.dot(np.transpose(x),np.subtract(h,y))
    return J , grad 
```

```python
theta_i=np.zeros(n+1)
test_theta = np.array([-24, 0.2, 0.2])
cost , grad =costFunc(x,y_input,test_theta)
```

```python
def gradient_descent(x, y, theta, alpha, num_iters):
    for i in range(num_iters):
        z=np.dot(x,theta)
        h_x=sigmoid(z)
        theta=theta-(alpha)*(1/m)*np.dot(np.transpose(x),np.subtract(h_x,y))
    return theta
```

```python
theta_i=np.zeros(n+1)
alpha=0.01
iteration=1000000
theta=gradient_descent(x,y_input,theta_i,alpha,iteration)
```

```python
def plotBoundary(theta):
    # Only need two points to define a line, so choose two endpoints
    plot_x = np.array([np.min(x[:, 1]) - 2, np.max(x[:, 1]) + 2])
    # Calculate the decision boundary line
    plot_y = (-1/theta[2]) * (theta[1]*plot_x + theta[0])
    plt.plot(plot_x, plot_y,label="Decision Boundary Boundary")
    plotData(data,x_input,y_input)
    
    
```

```python
plotBoundary(theta)
```

```python
initial_theta=np.zeros(n+1)
def cost_func(t):
    return costFunc(x,y_input,t)[0]


def grad_func(t):
    return costFunc(x,y_input,t)[1]

# Run fmin_bfgs to obtain the optimal theta
theta_opt, cost, *unused = opt.fmin_bfgs(f=cost_func, fprime=grad_func, x0=initial_theta, maxiter=400, full_output=True, disp=False)
```

```python
plotBoundary(theta_opt)
```
