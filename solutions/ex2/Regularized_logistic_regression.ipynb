```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
```

```python
def plotData(data,x,y):
    data_positive=data[data[:,-1]==1]
    data_negative=data[data[:,-1]==0]
    x_positive=data_positive[:,0:2]
    x_negative=data_negative[:,0:2]
    plt.scatter(x_positive[:,0],x_positive[:,1],color="k",marker="+",label="Admitted")
    plt.scatter(x_negative[:,0],x_negative[:,1],color="y",marker="o",label="Not admitted")
    plt.xlabel('Microchip Test 1')
    plt.ylabel('Microchip Test 2')
    plt.legend(loc="upper right") 
```

```python
data=np.loadtxt("ex2data2.txt",delimiter=",")
x_input=data[:,0:2]
y_input=data[:,-1]
m=y_input.shape[0]
n=x_input.shape[1]
```

```python
plotData(data,x_input,y_input)
```

```python
def map_feature(x1, x2):
    degree = 6

    x1 = x1.reshape((x1.size, 1))
    x2 = x2.reshape((x2.size, 1))
    result = np.ones(x1[:, 0].shape)

    for i in range(1, degree + 1):
        for j in range(0, i + 1):
            result = np.c_[result, (x1**(i-j)) * (x2**j)]

    return result
```

```python
x=map_feature(x_input[:,0],x_input[:,1])
```

```python
m=x.shape[0]
n=x.shape[1]
```

```python
def sigmoid(z):
    
    h=1/(1+np.exp(-z))
    return h
```

```python
def costFunctionReg(x,y,theta,l):
    (m,n)=x.shape
    z=np.dot(x,theta)
    
    h=sigmoid(z)
    
    log_h=np.log(h)
    h_1=np.subtract(1,h)
    log_h_1=np.log(h_1)
    y_1=np.subtract(1,y)
    
    J1=np.subtract(-y * log_h, y_1 * log_h_1)
    r=(l/(2*m))*(np.sum(np.square(theta[1:])))
    J=(1/m)*np.sum(J1)+r
    ######################################################## 
    #grad
    grad = np.zeros(theta.shape)
    E=np.subtract(h,y)
    grad=(1/m)*np.dot(x.T,E)
    grad[1:]=grad[1:]+(l/m)*theta[1:]
    return J , grad
```

```python
theta=np.zeros(n)
test_theta = np.ones(n)
```

```python
cost,grad=costFunctionReg(x,y_input,test_theta,1)
```
