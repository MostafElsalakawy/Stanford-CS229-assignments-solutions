```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.loadtxt("ex1data1.txt",delimiter=",",usecols=(0,1))
x_input=data[:,0]  #population in cities
y_input=data[:,1]  #profits
y=y_input          
m=x_input.shape[0]  # m number of trainig points
n=1          # n number of features
x_input=x_input.reshape((m,n))
y=y.reshape((m,1))
plt.figure()
plt.plot(x_input,y,"rx",markersize=10)
plt.xlabel('Population of City in 10,000s'); # Set the x axis label
plt.ylabel('Profit in $10,000s'); # Set the y axis label
plt.show()
```

```python
# add coloumn of ones to input data for intercept term
x=np.insert(x_input,0,np.ones(m),axis=1)
n=n+1    # number of features increament by 1 (x0)
theta=np.zeros((n,1))  # initialize theta to 0
```

```python
def compute_cost(X, y, theta):
    h_x=np.dot(x,theta)
    J_theta=0.5*(1/m)*np.sum(np.square(np.subtract(y,h_x)))
    return J_theta
    
    
```

```python
def gradient_descent(x, y, theta, alpha, num_iters):
    J_history=[]
    for i in range(num_iters):
        h_x=np.dot(x,theta)
        theta=theta-(alpha)*(1/m)*np.dot(np.transpose(x),np.subtract(h_x,y))
        J_history.append(compute_cost(x,y,theta))
    return theta , J_history
    
    
```

```python
compute_cost(x,y,theta)
```

```python
alpha=0.01
iteration=1500
theta,J_history=gradient_descent(x,y,theta,alpha,iteration)
```

```python
y=np.dot(x,theta)
```

```python
plt.figure()
plt.plot(x_input,y,"k",label="linear regression")
plt.plot(x_input,y_input,"rx",markersize=10,label="trainig data")
plt.xlabel('Population of City in 10,000s'); # Set the x axis label
plt.ylabel('Profit in $10,000s'); # Set the y axis label
plt.legend()
plt.show()
```
